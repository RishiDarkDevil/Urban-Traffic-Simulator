```python
import networkx as nx,pandas as pd, numpy as np,matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r'Calcutta/Calcutta_Edgelist.csv')
data
```

```python
# m=data.shape[0]
m=1000
g=np.array(data.iloc[20000:21000,[2,3,5]])   #store the 3rd and 4th column in an edge list np array
g
```

```python
# d=np.array([[1,2,['a']],[2,3,['b']]])
G=nx.DiGraph()
G.add_weighted_edges_from(g)
nx.draw(G,node_size=1,width=1,arrowsize=5)
```

```python
n=1   #number of connected components
c=nx.strongly_connected_components(G)  #c is a generator of sets of nodes of connected components 
lcomp=sorted(c,key=len,reverse=True)[:n]
# lcomp
```

```python
for i,comp in enumerate(lcomp):
    wcc=nx.subgraph(G,comp)    #returns the subgraph for the set of nodes in comp
    nx.draw(wcc,node_size=0.1,width=0.05,arrowsize=1)
    plt.savefig(f'Connected Components\scc{i+1}_{n}.png',dpi=10000)
    plt.show()
```
